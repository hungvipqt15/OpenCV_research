Template matching cho phép chúng ta phát hiện vật thể trong ảnh đầu vào bằng cách sử dụng một ảnh tham chiếu (template) chứa vật thể cần phát hiện. Cách thuật toán hoạt động:

- Trượt ảnh template trên ảnh đầu vào (giống tích chập 2D)
- Ảnh template sẽ được so sánh với cửa sổ trượt tương ứng trên ảnh đầu vào bằng một công thức.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('/content/drive/MyDrive/OpenCV/Photos/4.jpg')
plt.imshow(img[:,:,::-1])
```

```python
tiny_img = img[300:600,400:600]
plt.imshow(tiny_img)
```

```python
height, width, channels = tiny_img.shape
```

```python
res = cv2.matchTemplate(img, tiny_img, cv2.TM_CCOEFF_NORMED)
threshold = 0.90
loc = np.where(res>=threshold)
```

```python
for pt in zip(*loc[::-1]):
  cv2.rectangle(img, pt, (pt[0]+width, pt[1]+height),(255,0,0),2)

plt.imshow(img[:,:,::-1]); plt.title('Location in image');
plt.show()
```

Ưu điểm:

- Nhanh, đơn giản
- Không tốn công làm data

Nhược điểm

- Template phải rất giống với vật thể trong ảnh cả về kích thước độ nghiêng, ... Nếu khác biệt quá lớn sẽ không phát hiện được.
